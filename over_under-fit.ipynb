```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.linspace(0, 10, 1000)
ys = np.sin(2*np.pi*xs) + 1

x = 10 * np.random.random_sample(100)
y = np.sin(2*np.pi*x) + np.random.normal(1, 0.1)

plt.scatter(x, y)
plt.plot(xs, ys)
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
orders = [1, 3, 7, 11, 16, 20]
# orders = [1]
mse = []

models = [np.poly1d(np.polyfit(x, y, o)) for o in orders]


for cnt in range(len(orders)):
    ys_p = models[cnt](xs)
    y_p = models[cnt](x)
    mse.append(np.mean((y_p - y) ** 2))
    plt.plot(xs, ys_p, label='order = {}'.format(orders[cnt]))
    

plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()        


print('mse = {}'.format(mse))

plt.plot(orders, mse)
plt.xlabel('order')
plt.ylabel('mse')
plt.show()
```
